```python
#Import dependencies for training process
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, mean_squared_error, mean_absolute_error, r2_score
from xgboost import XGBClassifier, XGBRegressor
import numpy as np

```

<div style="background-color: #e6f2f2; padding: 15px; border-left: 5px solid #66b2b2; margin-bottom: 20px;">
  <h4 style="color: #333;">1. Encode Data For sklearn to understand</h4>
</div>


```python
# Encode target for classification
df['y_bin'] = df['y'].map({'yes': 1, 'no': 0})
```

```python
df.head() #check new last column "y_bin"
```

<div style="background-color: #e6f2f2; padding: 15px; border-left: 5px solid #66b2b2; margin-bottom: 20px;">
  <h4 style="color: #333;">2. Split Data into training and testing sets</h4>
</div>


```python
# Separate features and targets
X = df.drop(columns=['y', 'y_bin'])
y_class = df['y_bin']   # Classification target
y_reg = df['campaign']  # Regression target
```

```python
y_reg.values
```

```python
y_class.values
```

<div style="background-color: #e6f2f2; padding: 15px; border-left: 5px solid #66b2b2; margin-bottom: 20px;">
  <h4 style="color: #333;">3. Create a processing pipeline with StandardScaler() and OneHotEncoder()</h4>
</div>


```python
# Categorical and numerical columns
cat_cols = X.select_dtypes(include='object').columns.tolist()
num_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
```

```python
cat_cols #columns for categorical data
```

```python
num_cols #columns for numerical data
```

```python
# Handle "unknown" as a separate category in OneHotEncoder
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols), #scale numerical columns
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols) #encode categorical columns including "unknown" as a value
    ]
)
```

<div style="background-color: #e6f2f2; padding: 15px; border-left: 5px solid #66b2b2; margin-bottom: 20px;">
  <h2 style="color: #333;">--- Training Classification Models---</h2>
</div>


<div style="background-color: #e6f2f2; padding: 15px; border-left: 5px solid #66b2b2; margin-bottom: 20px;">
  <p style="color: #333;">In the following code, we tested <strong>three</strong> classification modelsâ€” <strong>Logistic Regression, Random Forest, and XGBoost</strong>-using a consistent <strong>pipeline setup</strong>. We first split the dataset into training and testing sets with stratification. Then, for each model, we created a <strong>pipeline</strong> that included preprocessing and the classifier, trained it on the training data, and made predictions on the test set. Finally, we evaluated each model using <strong>accuracy and F1 score,</strong> storing the results in a dictionary for comparison.</p>
</div>




```python
#Create a dictionary with models to be tested
models_class = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(random_state=42),
    'XGBoost': XGBClassifier(eval_metric='logloss', random_state=42)
}

#Split data into training and testing
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X, y_class, test_size=0.2, random_state=42, stratify=y_class)

#Create training loop
results_class = {}
for name, model in models_class.items():

    clf_pipeline = Pipeline(steps=[         # make use of Pipeline to train all models at once
        ('preprocessor', preprocessor),
        ('classifier', model)
    ])
    
    clf_pipeline.fit(X_train_c, y_train_c) # train models in pipeline

    preds = clf_pipeline.predict(X_test_c) # make some predications

    probs = clf_pipeline.predict_proba(X_test_c)[:, 1]

    results_class[name] = {   # create a dictionary with model name as key and metrics as values
        'Accuracy': accuracy_score(y_test_c, preds),
        'F1 Score': f1_score(y_test_c, preds),
        'ROC AUC': roc_auc_score(y_test_c, probs)
    }
```

<div style="background-color: #e6f2f2; padding: 15px; border-left: 5px solid #66b2b2; margin-bottom: 20px;">
  <h2 style="color: #333;">--- Training Regression Models ---</h2>
</div>


<div style="background-color: #e6f2f2; padding: 15px; border-left: 5px solid #66b2b2; margin-bottom: 20px;">
  <p style="color: #333;">In this code, we evaluated <strong>three</strong> regression modelsâ€”<strong>Linear Regression, Random Forest Regressor, and XGBoost Regressor</strong>â€”using a <strong>pipeline approach</strong>. We split the data into training and testing sets, and for each model, we built a <strong>pipeline</strong> that included <strong>preprocessing</strong> and the <strong>regressor</strong>. After training the models on the training data, we made predictions on the test set and calculated evaluation metrics including <strong>MAE, RMSE, and RÂ²</strong>, storing the results in a dictionary for easy comparison.</p>
</div>



```python
#Create a dictionary with models to be tested
models_reg = {
    'Linear Regression': LinearRegression(),
    'Random Forest Regressor': RandomForestRegressor(random_state=42),
    'XGBoost Regressor': XGBRegressor(random_state=42)
}

#Split data into training and testing
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(X, y_reg, test_size=0.2, random_state=42)

#Create training loop
results_reg = {}
for name, model in models_reg.items():

    reg_pipeline = Pipeline(steps=[  # make use of Pipeline to train all models at once
        ('preprocessor', preprocessor),
        ('regressor', model)
    ])

    reg_pipeline.fit(X_train_r, y_train_r)# train models in pipeline

    preds = reg_pipeline.predict(X_test_r) #make some predications

    results_reg[name] = {  # create a dictionary with model name as key and metrics as values
        'MAE': mean_absolute_error(y_test_r, preds),
        'RMSE': np.sqrt(mean_squared_error(y_test_r, preds)), #optionally you can use,  mean_squared_error(y_test_r, preds ,squared=False)
        'R2': r2_score(y_test_r, preds)
    }

```
